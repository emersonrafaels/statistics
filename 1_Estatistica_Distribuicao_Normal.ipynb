# Importing libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

# Global variables

```python
# Set: alpha
alpha = 0.05
```

# Generate random dataset

```python
data = np.random.normal(0, 1, 1000)
```

# Ploting a histogram

```python
plt.hist(data, bins=50, density=True)
plt.show()
```

# Ploting Q-Q Graph

```python
# Plota um gráfico Q-Q
stats.probplot(data, dist="norm", plot=plt)
plt.show()
```

# Test: Shapiro-Wilk

```python
stats, p = stats.shapiro(data)
print("Test Stats: ", stats)
print("p-value: ", p)
```

```python
if p > alpha:
    print("The dataset appears a normal distribution (p = {})".format(p))
else:
    print("The dataset not appears a normal distribution (p = {})".format(p))
```

The stats.shapiro function returns two values: the statistic test and the p-value. If the p-value greater than a significance value (usually 0.05), we can conclude that there not enough evidence to reject the null hypothesis that the dataset follow a normal distribution.

Otherwise, we can conclude that the dataset not follow a normal distribution.
